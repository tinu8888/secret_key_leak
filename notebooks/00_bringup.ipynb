```python
import matplotlib.pyplot as plt
from dlsca import seeds, capture

seeds.set_all(0)  # determinism
```

```python
# Detect the scope + target (claims USB, gated inside capture.connect()).
# capture.connect() surfaces an actionable error if the USB device can't be claimed
# (replug / release the device / check no other process holds it).
scope, target = capture.connect()
print(scope)
```

```python
# Capture one trace with a known test (key, plaintext) and plot it.
key = bytes(range(16))
plaintext = bytes(range(16))
wave, ciphertext = capture.capture_trace(key, plaintext)
print("ciphertext:", ciphertext.hex())

plt.figure()
plt.plot(wave)
plt.title("First captured power trace")
plt.xlabel("sample")
plt.ylabel("power (a.u.)")
plt.show()
```
